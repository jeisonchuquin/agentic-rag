Llamamos al retriever de la base vectorial.

```python
from src.infrastructure.retrievers import graph
from langchain.chains import GraphQAChain
from langchain_openai import ChatOpenAI
from src.config import OPENAI_API_KEY, opik_tracer
```

```python
llm = ChatOpenAI(
    # temperature=0,
    model="o3",
    openai_api_key=OPENAI_API_KEY,
    max_tokens=None,
    callbacks=[opik_tracer]
)
chain = GraphQAChain.from_llm(llm, graph=graph, verbose=True)
```

```python
QUESTION = 'notas para aprobar una materia'
retrieved_docs = chain.invoke(QUESTION)
```

```python
QUESTION = "¿Qué es la Usfq?"
retrieved_docs = chain.invoke(QUESTION)
```
